````python
import torch
import time
import re
import os
from collections import Counter
from datetime import datetime
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

MODEL_DIR = "/content/mT5_model"

# Auto-detect if model files are in a subfolder
if "config.json" not in os.listdir(MODEL_DIR):
    for item in os.listdir(MODEL_DIR):
        subpath = os.path.join(MODEL_DIR, item)
        if os.path.isdir(subpath) and "config.json" in os.listdir(subpath):
            MODEL_DIR = subpath
            break
print(f"📂 Model path: {MODEL_DIR}")
print(f"📄 Files: {os.listdir(MODEL_DIR)}")

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🖥️ Device: {device.upper()}")

print("⏳ Loading tokenizer & model...")
tokenizer = AutoTokenizer.from_pretrained(MODEL_DIR, legacy=True)
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_DIR).to(device)
model.eval()
print(f"✅ Model loaded! ({sum(p.numel() for p in model.parameters())/1e6:.0f}M params)")

LANGUAGES = {
    "hindi": ("Hindi", "हिन्दी", 0x0900, 0x097F),
    "bengali": ("Bengali", "বাংলা", 0x0980, 0x09FF),
    "tamil": ("Tamil", "தமிழ்", 0x0B80, 0x0BFF),
    "telugu": ("Telugu", "తెలుగు", 0x0C00, 0x0C7F),
    "marathi": ("Marathi", "मराठी", 0x0900, 0x097F),
    "gujarati": ("Gujarati", "ગુજરાતી", 0x0A80, 0x0AFF),
    "punjabi": ("Punjabi", "ਪੰਜਾਬੀ", 0x0A00, 0x0A7F),
    "urdu": ("Urdu", "اردو", 0x0600, 0x06FF),
    "kannada": ("Kannada", "ಕನ್ನಡ", 0x0C80, 0x0CFF),
    "malayalam": ("Malayalam", "മലయാളം", 0x0D00, 0x0D7F),
}

def detect_language(text):
    counts = {}
    for ch in text:
        cp = ord(ch)
        for lang, (_, _, lo, hi) in LANGUAGES.items():
            if lo <= cp <= hi:
                counts[lang] = counts.get(lang, 0) + 1
    if not counts:
        return "English", "English"
    key = max(counts, key=counts.get)
    return LANGUAGES[key][0], LANGUAGES[key][1]

def summarize(text, max_length=150, min_length=30):
    try:
        text = re.sub(r'\s+', ' ', text).strip()
        lang_name, lang_native = detect_language(text)

        t0 = time.time()
        inputs = tokenizer(
            text, return_tensors="pt",
            padding="max_length", truncation=True, max_length=512
        ).to(device)

        input_toks = (inputs["input_ids"] != tokenizer.pad_token_id).sum().item()

        with torch.no_grad():
            out = model.generate(
                input_ids=inputs["input_ids"],
                attention_mask=inputs["attention_mask"],
                max_length=max_length,
                min_length=min_length,
                num_beams=4,
                length_penalty=2.0,
                no_repeat_ngram_size=3,
                early_stopping=True
            )

        summary = tokenizer.decode(out[0], skip_special_tokens=True, clean_up_tokenization_spaces=False)
        latency = round(time.time() - t0, 2)
        out_toks = (out[0] != tokenizer.pad_token_id).sum().item()
        compression = round(len(text) / max(len(summary), 1), 2)

        return {
            "summary": summary, "language": lang_name, "native": lang_native,
            "input_tokens": input_toks, "output_tokens": out_toks,
            "compression": compression, "latency": latency, "error": None
        }
    except Exception as e:
        return {"error": str(e)}

print("\n🧪 Running smoke test...")
test = summarize("भारत एक महान देश है। यहां विभिन्न संस्कृतियां और भाषाएं हैं। भारत में अनेक त्योहार मनाए जाते हैं।")
if test.get("error"):
    print(f"❌ Error: {test['error']}")
else:
    print(f"✅ Smoke test passed! Summary: {test['summary']}")
print("🚀 Summarizer ready!")
````

````python
import gradio as gr

history_log = []

def chat(user_msg, chat_history, max_len, min_len):
    msg = user_msg.strip()

    if msg.lower() in {"hi", "hello", "help", "namaste", "start"}:
        reply = (
            "🙏 **Namaste! Indian Language News Summarizer**\n\n"
            "**Paste any news article** in Hindi, Bengali, Tamil, Telugu, "
            "Marathi, Gujarati, Punjabi, Urdu, Kannada, or Malayalam.\n\n"
            "I'll auto-detect the language and summarize it! 📰"
        )
        chat_history.append((msg, reply))
        return "", chat_history

    if msg.lower() == "history":
        if not history_log:
            reply = "📭 No history yet."
        else:
            rows = "\n".join(f"  {h['language']:10} | {h['compression']}x | {h['latency']}s" for h in history_log[-10:])
            reply = f"📜 **Last {min(10, len(history_log))} runs:**\n```\n  Language   | Compress | Latency\n{rows}\n```"
        chat_history.append((msg, reply))
        return "", chat_history

    if len(msg) < 50:
        chat_history.append((msg, "⚠️ Too short! Paste a full news article."))
        return "", chat_history

    r = summarize(msg, max_length=int(max_len), min_length=int(min_len))

    if r.get("error"):
        reply = f"❌ Error: {r['error']}"
    else:
        history_log.append(r)
        reply = (
            f"## 📝 Summary ({r['language']} {r['native']})\n\n"
            f"> {r['summary']}\n\n"
            f"---\n"
            f"| Metric | Value |\n|---|---|\n"
            f"| 🌐 Language | {r['language']} |\n"
            f"| 📄 Input | {r['input_tokens']} tokens |\n"
            f"| 📝 Output | {r['output_tokens']} tokens |\n"
            f"| 📐 Compression | {r['compression']}x |\n"
            f"| ⏱️ Latency | {r['latency']}s |"
        )

    chat_history.append((msg, reply))
    return "", chat_history

def clear_chat():
    return [], ""

with gr.Blocks(title="🇮🇳 Indian News Summarizer") as demo:
    gr.Markdown("# 🇮🇳 Indian Language News Summarization Chatbot\n*Powered by mT5 — BTech Project (10 Credits)*\n\n---")

    with gr.Row():
        with gr.Column(scale=1):
            max_sl = gr.Slider(50, 300, 150, step=10, label="Max Summary Length")
            min_sl = gr.Slider(10, 100, 30, step=5, label="Min Summary Length")
            clear_btn = gr.Button("🗑️ Clear", variant="secondary")
            gr.Markdown("**Commands:** `help` · `history`\n\n**Languages:** Hindi · Bengali · Tamil · Telugu · Marathi · Gujarati · Punjabi · Urdu · Kannada · Malayalam")

        with gr.Column(scale=3):
            chatbot = gr.Chatbot(height=520, label="💬 Summarizer")
            msg_box = gr.Textbox(placeholder="Paste a news article...", label="📰 Input", lines=3)
            send_btn = gr.Button("📤 Summarize", variant="primary")

    send_btn.click(chat, [msg_box, chatbot, max_sl, min_sl], [msg_box, chatbot])
    msg_box.submit(chat, [msg_box, chatbot, max_sl, min_sl], [msg_box, chatbot])
    clear_btn.click(clear_chat, outputs=[chatbot, msg_box])

demo.launch(share=True)
````
